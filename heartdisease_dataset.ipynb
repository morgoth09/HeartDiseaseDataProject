```python
df_hd = df[df['Topic'] != 'All stroke']
df_hd['Topic'].unique()
```

```python
df_hd_sex = df_hd[df_hd['StratificationCategory3'] == 'Sex']
df_hd_sex['Stratification3'].value_counts()
```

```python
avg_rates = (
    df_hd_sex.groupby('Stratification3')['Data_Value']
    .mean()
    .round(1)
    .reset_index()
)
avg_rates
```

Percent difference of death rates between men and women

```python
male_rate = avg_rates.loc[avg_rates['Stratification3'] == 'Men', 'Data_Value'].values[0]
female_rate = avg_rates.loc[avg_rates['Stratification3'] == 'Women', 'Data_Value'].values[0]

percent_diff = ((male_rate - female_rate) / female_rate) * 100

percent_diff_rounded = round(percent_diff, 1)

percent_diff_rounded.item()
```

Data Chart

```python
plt.figure(figsize=(8,5))
plt.bar(
    ['Men', 'Women'],
    [male_rate, female_rate],
    color = ["#66B2FF", "#FF9999"] 
)

plt.title('Average Heart Disease Death Rate: Men vs Women')
plt.ylabel('Death Rate per 100,00')
plt.xlabel('Sex')
plt.ylim(0, max(male_rate, female_rate) * 1.1) #Goes above max value to add a gap at the top and make it look nicer

#add numbers on the top of the bars
for i, v in enumerate([male_rate, female_rate]):
    plt.text(i, v + 10, f"{v:.1f}", ha='center', fontweight = 'bold')

    #Add precent difference 
    plt.text(
        0,
        male_rate + 40, #y coordinate
        f"Male death rate is {percent_diff:.1f}% higher than women.",
        ha='left',
        fontweight='bold',
        

    )

plt.show()
```
